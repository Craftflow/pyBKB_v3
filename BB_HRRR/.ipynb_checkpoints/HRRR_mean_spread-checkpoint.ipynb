# Mean HRRR Model Spread

The model spread is the standard deviation between all the members. The mean model spread, however, is the square root of the mean variance.

```python
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

import sys
sys.path.append('/uufs/chpc.utah.edu/common/home/u0553130/pyBKB_v3')
sys.path.append('B:\pyBKB_v3')
from BB_HRRR.HRRR_Spread import mean_spread
from BB_datetimes.range import range_dates
```

```python
hour = 12
sDATE = datetime(2018, 9, 1, hour)
eDATE = datetime(2018, 9, 5, hour)
DATES = range_dates(sDATE, eDATE, DAYS=1)

fxx = range(0,4)
```

```python
DATES
```

```python
AVG_SPREAD = mean_spread(DATES, variable='TMP:2 m', fxx=fxx)
```

```python
plt.figure(figsize=(12,8))
plt.pcolormesh(AVG_SPREAD, cmap='magma')
plt.title('HRRR Model Spread', fontweight='bold', loc='left')
plt.title('sDATE: %s\n eDATE: %s' % (sDATE.strftime('%Y %b %H:%M UTC'),
                                     eDATE.strftime('%Y %b %H:%M UTC')), loc='right')
plt.title('Hour: %s' % hour)
plt.colorbar()
```
